```python
from classifier import get_data, get_model, predict
import matplotlib.pyplot as plt
import numpy as np
import collections
```

```python
X, Y, train_X, valid_X, train_Y, valid_Y = get_data("adventure-horror")
model = get_model("sgd")
```

```python
predicted_classes = predict(model, valid_X, valid_Y)
```

```python
print("Predicted", collections.Counter(predicted_classes))
print("Valid", collections.Counter(valid_Y))
```

```python
count = np.zeros((2,2), dtype='int16')
for i in range(len(predicted_classes)):
    if predicted_classes[i] != valid_Y[i]:
        count[predicted_classes[i], 0] += 1
    else:
        count[predicted_classes[i], 1] += 1

for i in range(len(count)):
    print("predicted class", i, "false", count[i, 0], "true", count[i, 1])
```
