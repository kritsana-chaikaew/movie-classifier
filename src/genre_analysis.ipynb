```python
import numpy as np
```

```python
data = np.genfromtxt('../genre_list/all.txt', delimiter=',', dtype='U')
data
```

```python
genre_list = data[:, 1]
genre_list
```

```python
genre_list.shape
```

```python
genre_list = np.unique(genre_list)
genre_list.shape
```

```python
genres = []
for i, x in zip(range(len(genre_list)), genre_list):
    genres.insert(i, x.split(':'))
genres
```

```python
union_genres = []
count = -1
temp = ""
for i in genres:
    if i[0] != temp:
        temp = i[0]
        count += 1
        union_genres.insert(count, set(i))
    else:
        union_genres[count] = set(union_genres[count]).union(i)
union_genres
```
